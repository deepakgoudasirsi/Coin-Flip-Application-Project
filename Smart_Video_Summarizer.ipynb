```python
from langchain.document_loaders import YoutubeLoader
from IPython.display import display, FileLink

# Function to fetch the transcript from a YouTube video
def fetch_transcript(video_url):
    loader = YoutubeLoader.from_youtube_url(video_url)
    transcript = loader.load()  # Fetch the transcript
    return transcript

# Save transcript to a text file
def save_transcript_to_file(transcript, filename='transcript.txt'):
    # Combine all transcript content into a single string
    full_text = "\n".join([doc.page_content for doc in transcript])

    # Write to a text file
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(full_text)
    print(f"Transcript saved as {filename}")

    return filename

# Download function (useful in Jupyter or web-based environments)
def download_transcript(filename):
    return FileLink(filename)

# Main flow
try:
    # Ask user to input the YouTube URL
    video_url = input("Please enter the YouTube video URL: ")

    # Fetch and display the transcript
    transcript = fetch_transcript(video_url)
    for idx, doc in enumerate(transcript):
        print(f"Part {idx + 1}:\n{doc.page_content}\n")

    # Save transcript to a file
    filename = save_transcript_to_file(transcript)

    # Provide a download link (Jupyter/IPython Notebook)
    download_link = download_transcript(filename)
    display(download_link)

except Exception as e:
    print(f"An error occurred: {e}")
```


# **Text to Audio Converter**:
    - Implement a feature that converts generated text summaries into audio format, making it accessible for users who prefer listening.

```python
from langchain.document_loaders import YoutubeLoader
from gtts import gTTS
import os
from IPython.display import Audio  # Use IPython display for audio playback

# Function to fetch the transcript from a YouTube video
def fetch_transcript(video_url):
    loader = YoutubeLoader.from_youtube_url(video_url)
    transcript = loader.load()
    return transcript

# Function to convert transcript to audio
def convert_text_to_audio(transcript, audio_file_name='transcript.mp3'):
    # Extract text from Document objects
    texts = [doc.page_content for doc in transcript]  # Extracting the content
    full_text = " ".join(texts)  # Combine all text into one string

    # Create audio from the text
    tts = gTTS(text=full_text, lang='en')
    tts.save(audio_file_name)  # Save the audio file
    print(f"Audio saved as {audio_file_name}")

    # Check if the file exists
    if os.path.exists(audio_file_name):
        print("Audio file exists.")
    else:
        print("Error: Audio file was not created.")

    return audio_file_name  # Return the name of the saved audio file

# Function to play the audio file with a play button
def play_audio(audio_file_name):
    # Display audio player with controls in Jupyter notebook
    return Audio(audio_file_name, autoplay=False)  # Do not autoplay; user will press play

# Main flow
try:
    # Ask user to input the YouTube URL
    video_url = input("Please enter the YouTube video URL: ")

    # Step 1: Fetch the transcript using the input URL
    transcript = fetch_transcript(video_url)

    # Step 2: Convert the transcript to audio
    audio_file_name = convert_text_to_audio(transcript)

    # Step 3: Display the audio player with play button
    audio_player = play_audio(audio_file_name)
    display(audio_player)  # Display the audio player with controls (play/pause)

except Exception as e:
    print(f"An error occurred: {e}")
```


# Multi-Language Support:
    - Incorporate support for multiple languages to cater to a diverse user base, enabling transcription and summarization in various languages.

```python
from langchain.document_loaders import YoutubeLoader
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from deep_translator import GoogleTranslator

# Function to fetch the transcript from a YouTube video
def fetch_transcript(video_url):
    loader = YoutubeLoader.from_youtube_url(video_url)
    transcript = loader.load()  # Fetch the transcript
    return transcript

# Function to summarize the texts from the transcript
def summarize_texts(transcript):
    model_repo = 'gpt2'  # You can choose a different model
    tokenizer = AutoTokenizer.from_pretrained(model_repo)
    model = AutoModelForCausalLM.from_pretrained(model_repo)

    # Set up the text generation pipeline
    pipe = pipeline(
        task="text-generation",
        model=model,
        tokenizer=tokenizer,
        max_length=150,          # Maximum length of the input
        temperature=0.7,         # Sampling temperature
        top_p=0.9,               # Nucleus sampling
        repetition_penalty=1.2,  # Penalty for repetition
        pad_token_id=tokenizer.eos_token_id,
        truncation=True           # Enable truncation
    )

    summaries = []
    for doc in transcript:
        # Generate summary for each piece of text in the transcript
        summary = pipe(
            doc.page_content,
            max_new_tokens=50,  # Set this to control the length of the output
            num_return_sequences=1,
            truncation=True
        )[0]['generated_text']
        summaries.append(summary.strip())  # Strip any leading/trailing whitespace

    return summaries

# Function to translate text into the target language
def translate_text(text, target_language):
    translation = GoogleTranslator(source='auto', target=target_language).translate(text)
    return translation  # Return the translated text

# Function to save the translated summaries to a text file
def save_translated_summaries(translated_summaries, filename="translated_summaries.txt"):
    try:
        with open(filename, "w", encoding="utf-8") as file:
            for idx, summary in enumerate(translated_summaries):
                file.write(f"Summary {idx + 1}:\n{'-'*40}\n{summary}\n{'-'*40}\n\n")
        print(f"Summaries have been saved to {filename}")
    except Exception as e:
        print(f"Error saving the file: {e}")

# Main execution flow
try:
    # Ask user for the YouTube video URL and the target language
    video_url = input("Please enter the YouTube video URL: ")
    target_lang = input("Enter the target language (e.g., 'es' for Spanish, 'fr' for French, 'kn' for Kannada, 'hi' for Hindi etc.): ")

    # Fetch the transcript from the video
    transcript = fetch_transcript(video_url)

    # Generate summaries for the transcript
    summaries = summarize_texts(transcript)

    # Translate the summaries into the target language
    translated_summaries = [translate_text(summary, target_lang) for summary in summaries]

    # Print the translated summaries in a clear format
    print("\n--- Translated Summaries of the Video ---\n")
    for idx, translated_summary in enumerate(translated_summaries):
        print(f"Summary {idx + 1} in {target_lang}:\n{'-'*40}\n{translated_summary}\n{'-'*40}\n")

    # Save translated summaries to a file
    save_translated_summaries(translated_summaries)

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from langchain.document_loaders import YoutubeLoader
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import textstat
from textblob import TextBlob  # Importing TextBlob for sentiment analysis
from datetime import datetime

# Function to fetch the transcript from a YouTube video
def fetch_transcript(video_url):
    loader = YoutubeLoader.from_youtube_url(video_url)
    transcript = loader.load()  # Fetch the transcript
    return transcript

# Function to summarize the texts from the transcript
def summarize_texts(transcript):
    model_repo = 'gpt2'  # You can choose a different model
    tokenizer = AutoTokenizer.from_pretrained(model_repo)
    model = AutoModelForCausalLM.from_pretrained(model_repo)

    # Set up the text generation pipeline
    pipe = pipeline(
        task="text-generation",
        model=model,
        tokenizer=tokenizer,
        max_length=150,          # Maximum length of the input
        temperature=0.7,         # Sampling temperature
        top_p=0.9,               # Nucleus sampling
        repetition_penalty=1.2,  # Penalty for repetition
        pad_token_id=tokenizer.eos_token_id,
        truncation=True           # Enable truncation
    )

    summaries = []
    for doc in transcript:
        # Generate summary for each piece of text in the transcript
        summary = pipe(
            doc.page_content,
            max_new_tokens=50,  # Set this to control the length of the output
            num_return_sequences=1,
            truncation=True
        )[0]['generated_text']
        summaries.append(summary.strip())  # Strip any leading/trailing whitespace

    return summaries

# Function to analyze summary effectiveness
def analyze_summary_effectiveness(original_text, summaries):
    readability_scores = [textstat.flesch_reading_ease(summary) for summary in summaries]

    # Using TextBlob for sentiment analysis
    sentiment_scores = [TextBlob(summary).sentiment.polarity for summary in summaries]

    original_readability = textstat.flesch_reading_ease(original_text)
    effectiveness = {
        "original_readability": original_readability,
        "summaries_readability": readability_scores,
        "summaries_sentiment": sentiment_scores
    }

    return effectiveness

# Function to visualize analytics
def visualize_analytics(effectiveness, engagement_data):
    # Prepare the data for visualization
    summary_index = list(range(1, len(effectiveness['summaries_readability']) + 1))

    # Creating two separate subplots: One for readability score comparison, and the other for engagement data
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))  # 1 row, 2 columns

    # Plotting Readability Score Comparison
    axes[0].bar(summary_index, effectiveness['summaries_readability'], color='skyblue', label='Summary Readability')
    axes[0].axhline(y=effectiveness['original_readability'], color='r', linestyle='--', label='Original Readability')
    axes[0].set_xlabel('Summary Number')
    axes[0].set_ylabel('Readability Score')
    axes[0].set_title('Readability Score Comparison')
    axes[0].legend()

    # Engagement Data Visualization
    engagement_df = pd.DataFrame(engagement_data)
    engagement_df.set_index('summary_number', inplace=True)
    engagement_df.plot(kind='bar', ax=axes[1], figsize=(10, 6), legend=False)
    axes[1].set_title('Summary Engagement Analytics')
    axes[1].set_xlabel('Summary Number')
    axes[1].set_ylabel('User Engagement')

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()

# Main execution flow
try:
    # Ask user for the YouTube video URL
    video_url = input("Please enter the YouTube video URL: ")

    # Fetch the transcript from the video
    transcript = fetch_transcript(video_url)

    # Generate summaries for the transcript
    summaries = summarize_texts(transcript)

    # For simplicity, assuming the original text is the first item in the transcript for readability comparison
    original_text = transcript[0].page_content

    # Analyze summary effectiveness (readability and sentiment)
    effectiveness = analyze_summary_effectiveness(original_text, summaries)

    # Simulate user engagement data (e.g., time spent on each summary and views)
    engagement_data = [
        {'summary_number': i+1, 'views': 10*(i+1), 'time_spent': 3*(i+1)}  # Simulated values
        for i in range(len(summaries))
    ]

    # Visualize the analytics
    visualize_analytics(effectiveness, engagement_data)

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from langchain.document_loaders import YoutubeLoader
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import textstat
from textblob import TextBlob  # Importing TextBlob for sentiment analysis
from datetime import datetime

# Function to fetch the transcript from a YouTube video
def fetch_transcript(video_url):
    loader = YoutubeLoader.from_youtube_url(video_url)
    try:
        transcript = loader.load()  # Fetch the transcript
        if not transcript:
            raise ValueError("No transcript found for the provided video URL.")
        return transcript
    except Exception as e:
        raise ValueError(f"An error occurred while fetching the transcript: {e}")

# Function to summarize the texts from the transcript
def summarize_texts(transcript):
    model_repo = 'gpt2'  # You can choose a different model
    tokenizer = AutoTokenizer.from_pretrained(model_repo)
    model = AutoModelForCausalLM.from_pretrained(model_repo)

    # Set up the text generation pipeline
    pipe = pipeline(
        task="text-generation",
        model=model,
        tokenizer=tokenizer,
        max_length=150,          # Maximum length of the input
        temperature=0.7,         # Sampling temperature
        top_p=0.9,               # Nucleus sampling
        repetition_penalty=1.2,  # Penalty for repetition
        pad_token_id=tokenizer.eos_token_id,
        truncation=True           # Enable truncation
    )

    summaries = []
    for doc in transcript:
        # Check if 'page_content' exists in the document
        if hasattr(doc, 'page_content'):
            summary = pipe(
                doc.page_content,
                max_new_tokens=50,  # Set this to control the length of the output
                num_return_sequences=1,
                truncation=True
            )[0]['generated_text']
            summaries.append(summary.strip())  # Strip any leading/trailing whitespace
        else:
            print(f"Skipping document with missing content: {doc}")

    if not summaries:
        raise ValueError("No valid summaries generated.")

    return summaries

# Function to analyze summary effectiveness
def analyze_summary_effectiveness(original_text, summaries):
    readability_scores = [textstat.flesch_reading_ease(summary) for summary in summaries]

    # Sentiment analysis using TextBlob
    sentiment_scores = [TextBlob(summary).sentiment.polarity for summary in summaries]

    original_readability = textstat.flesch_reading_ease(original_text)
    effectiveness = {
        "original_readability": original_readability,
        "summaries_readability": readability_scores,
        "summaries_sentiment": sentiment_scores
    }

    return effectiveness

# Function to visualize analytics with sentiment scores
def visualize_analytics(effectiveness, engagement_data):
    # Prepare the data for visualization
    summary_index = list(range(1, len(effectiveness['summaries_readability']) + 1))

    # Creating two separate subplots: One for readability score comparison, and the other for engagement and sentiment data
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # 1 row, 3 columns

    # Plotting Readability Score Comparison
    axes[0].bar(summary_index, effectiveness['summaries_readability'], color='skyblue', label='Summary Readability')
    axes[0].axhline(y=effectiveness['original_readability'], color='r', linestyle='--', label='Original Readability')
    axes[0].set_xlabel('Summary Number')
    axes[0].set_ylabel('Readability Score')
    axes[0].set_title('Readability Score Comparison')
    axes[0].legend()

    # Plotting Sentiment Scores
    axes[1].bar(summary_index, effectiveness['summaries_sentiment'], color='lightgreen', label='Sentiment Scores')
    axes[1].set_xlabel('Summary Number')
    axes[1].set_ylabel('Sentiment Score')
    axes[1].set_title('Sentiment Analysis of Summaries')

    # Engagement Data Visualization
    engagement_df = pd.DataFrame(engagement_data)
    engagement_df.set_index('summary_number', inplace=True)
    engagement_df.plot(kind='bar', ax=axes[2], figsize=(10, 6), legend=False)
    axes[2].set_title('Summary Engagement Analytics')
    axes[2].set_xlabel('Summary Number')
    axes[2].set_ylabel('User Engagement')

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()

# Main execution flow
try:
    # Ask user for the YouTube video URL
    video_url = input("Please enter the YouTube video URL: ")

    # Fetch the transcript from the video
    transcript = fetch_transcript(video_url)

    # Check if transcript is empty
    if not transcript:
        raise ValueError("The transcript is empty or couldn't be fetched. Please ensure the video has a transcript.")

    # Generate summaries for the transcript
    summaries = summarize_texts(transcript)

    # For simplicity, assuming the original text is the first item in the transcript for readability comparison
    original_text = transcript[0].page_content if transcript else ""

    # Analyze summary effectiveness (readability, sentiment)
    effectiveness = analyze_summary_effectiveness(original_text, summaries)

    # Simulate user engagement data (e.g., time spent on each summary and views)
    engagement_data = [
        {'summary_number': i+1, 'views': 10*(i+1), 'time_spent': 3*(i+1)}  # Simulated values
        for i in range(len(summaries))
    ]

    # Visualize the analytics with sentiment analysis
    visualize_analytics(effectiveness, engagement_data)

except Exception as e:
    print(f"An error occurred: {e}")
```


#Time-Based Summaries:
    - Offer the ability to generate summaries based on specific time intervals, allowing users to navigate quickly through long videos.

```python
import yt_dlp
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled, NoTranscriptFound
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
import pandas as pd

# Step 1: Download YouTube transcript (if available)
def download_transcript(youtube_url):
    try:
        # Extract YouTube video ID from URL
        video_id = youtube_url.split("v=")[1].split("&")[0] if "v=" in youtube_url else youtube_url.split("/")[-1]

        # Fetch transcript using YouTube Transcript API
        transcript = YouTubeTranscriptApi.get_transcript(video_id)

        # If transcript is available, return it
        return transcript
    except (TranscriptsDisabled, NoTranscriptFound) as e:
        print(f"Transcript not available for this video: {e}")
        return None
    except Exception as e:
        print(f"Error downloading transcript: {e}")
        return None

# Step 2: Summarize transcript based on time intervals
def summarize_transcript_by_time(transcript, time_interval=60):
    # Load Hugging Face BART model for summarization
    model_repo = 'facebook/bart-large-cnn'
    tokenizer = AutoTokenizer.from_pretrained(model_repo)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_repo)
    summarizer = pipeline('summarization', model=model, tokenizer=tokenizer)

    summaries = []

    # Break transcript into time intervals and generate summaries
    for i in range(0, len(transcript), time_interval):
        # Create a chunk of text based on the time interval
        interval_text = " ".join([entry['text'] for entry in transcript[i:i+time_interval]])

        # Summarize the chunk
        summary = summarizer(interval_text, max_length=200, min_length=50, do_sample=False)[0]['summary_text']

        # Get start time for the interval
        start_time = transcript[i]['start']

        summaries.append({
            'start_time': start_time,
            'summary': summary.strip()
        })

    return summaries

# Step 3: Display summaries with time
def display_summaries(summaries):
    # Convert to pandas DataFrame for easy viewing
    df = pd.DataFrame(summaries)
    print(df)

# Step 4: Main function to run the workflow
def main():
    # Ask the user to input the YouTube URL
    youtube_url = input("Please enter the YouTube URL: ")

    transcript = download_transcript(youtube_url)

    if transcript:
        # Generate summaries based on time intervals (e.g., every 60 seconds)
        summaries = summarize_transcript_by_time(transcript, time_interval=60)

        # Display the summaries with start times
        display_summaries(summaries)
    else:
        print("Transcript not available for this video.")

# Example usage
if __name__ == "__main__":
    main()
```

```python
import yt_dlp
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled, NoTranscriptFound
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
from gtts import gTTS  # Import gTTS for text-to-speech functionality
import moviepy.editor as mp
import os
import pandas as pd

# Step 1: Download YouTube transcript (if available)
def download_transcript(youtube_url):
    try:
        # Extract YouTube video ID from URL
        video_id = youtube_url.split("v=")[1].split("&")[0] if "v=" in youtube_url else youtube_url.split("/")[-1]

        # Fetch transcript using YouTube Transcript API
        transcript = YouTubeTranscriptApi.get_transcript(video_id)

        # If transcript is available, return it
        return transcript
    except (TranscriptsDisabled, NoTranscriptFound) as e:
        print(f"Transcript not available for this video: {e}")
        return None
    except Exception as e:
        print(f"Error downloading transcript: {e}")
        return None

# Step 2: Summarize transcript based on time intervals
def summarize_transcript_by_time(transcript, time_interval=60):
    # Load Hugging Face BART model for summarization
    model_repo = 'facebook/bart-large-cnn'
    tokenizer = AutoTokenizer.from_pretrained(model_repo)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_repo)
    summarizer = pipeline('summarization', model=model, tokenizer=tokenizer)

    summaries = []

    # Break transcript into time intervals and generate summaries
    for i in range(0, len(transcript), time_interval):
        # Create a chunk of text based on the time interval
        interval_text = " ".join([entry['text'] for entry in transcript[i:i+time_interval]])

        # Summarize the chunk
        summary = summarizer(interval_text, max_length=200, min_length=50, do_sample=False)[0]['summary_text']

        # Get start time for the interval
        start_time = transcript[i]['start']

        summaries.append({
            'start_time': start_time,
            'summary': summary.strip()
        })

    return summaries

# Step 3: Convert text summary to speech (audio)
def text_to_speech(summary, filename="summary_audio.mp3"):
    # Use Google TTS (gTTS) for text-to-speech conversion
    tts = gTTS(text=summary, lang='en')
    tts.save(filename)  # Save the speech to an audio file
    print(f"Audio summary saved as {filename}")

# Step 4: Extract video clip corresponding to the summary's time range
def extract_video_clip(youtube_url, start_time, end_time, output_filename="summary_clip.mp4"):
    # Set up video download options using yt-dlp
    ydl_opts = {
        'format': 'bestvideo+bestaudio',
        'outtmpl': 'downloaded_video.%(ext)s',
        'postprocessors': [{
            'key': 'FFmpegVideoConvertor',
            'preferedformat': 'mp4'
        }],
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([youtube_url])

    # Load the downloaded video using moviepy
    video = mp.VideoFileClip("downloaded_video.mp4")

    # Cut the video clip based on start and end times
    video_clip = video.subclip(start_time, end_time)
    video_clip.write_videofile(output_filename, codec='libx264')
    print(f"Video clip saved as {output_filename}")

    # Optionally, delete the full downloaded video to save space
    os.remove("downloaded_video.mp4")

# Step 5: Display summaries with time
def display_summaries(summaries):
    # Convert to pandas DataFrame for easy viewing
    df = pd.DataFrame(summaries)
    print(df)

# Step 6: Main function to run the workflow
def main():
    # Ask the user to input the YouTube URL
    youtube_url = input("Please enter the YouTube URL: ")

    transcript = download_transcript(youtube_url)

    if transcript:
        # Generate summaries based on time intervals (e.g., every 60 seconds)
        summaries = summarize_transcript_by_time(transcript, time_interval=60)

        # Display the summaries with start times
        display_summaries(summaries)

        # Ask user for the summary option to download
        summary_choice = input("Choose summary format (text, audio, video): ").strip().lower()

        if summary_choice == "text":
            # Save the summary as a text file
            with open("summary.txt", "w") as f:
                for summary in summaries:
                    f.write(f"Start Time: {summary['start_time']} seconds\n")
                    f.write(f"Summary: {summary['summary']}\n\n")
            print("Text summary saved as 'summary.txt'.")

        elif summary_choice == "audio":
            # Combine all summaries into one audio file
            full_summary = " ".join([summary['summary'] for summary in summaries])
            text_to_speech(full_summary)

        elif summary_choice == "video":
            # Ask user for a time range to extract from the video
            start_time = float(input("Enter the start time for the video clip (in seconds): "))
            end_time = float(input("Enter the end time for the video clip (in seconds): "))
            extract_video_clip(youtube_url, start_time, end_time)
        else:
            print("Invalid choice. Please choose 'text', 'audio', or 'video'.")
    else:
        print("Transcript not available for this video.")

# Example usage
if __name__ == "__main__":
    main()
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
from transformers import pipeline
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Download NLTK data
nltk.download('stopwords')
nltk.download('punkt')

# Initialize zero-shot classification pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Define potential topics
candidate_topics = [
    "technology", "education", "health", "science", "entertainment",
    "sports", "politics", "business", "travel", "history", "literature",
    "music", "art", "machine learning", "artificial intelligence",
    "software development", "cybersecurity", "environment", "psychology",
    "economics", "engineering", "food", "culture", "social media", "self-help"
]

# Step 1: Fetch and Clean Transcript
def get_cleaned_transcript(youtube_url):
    try:
        video_id = youtube_url.split("v=")[1].split("&")[0] if "v=" in youtube_url else youtube_url.split("/")[-1]
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        full_text = " ".join([entry['text'] for entry in transcript])

        # Clean text: tokenize and remove stopwords
        words = word_tokenize(full_text.lower())
        filtered_text = " ".join([word for word in words if word.isalnum() and word not in stopwords.words("english")])

        return filtered_text
    except Exception as e:
        print("Transcript could not be retrieved:", e)
        return None

# Step 2: Detect Topics and Generate Tags
def detect_topics_and_generate_tags(text):
    # Use zero-shot classifier for topic detection
    results = classifier(text, candidate_labels=candidate_topics, multi_class=True)

    # Select topics with high confidence as tags
    tags = [label for label, score in zip(results['labels'], results['scores']) if score > 0.5]
    return tags

# Main function to detect topics and tags
def generate_video_tags(youtube_url):
    # Get and clean transcript
    transcript_text = get_cleaned_transcript(youtube_url)
    if not transcript_text:
        return None

    # Detect topics
    tags = detect_topics_and_generate_tags(transcript_text)
    print("Detected Tags for the Video:", tags)

# Example usage
youtube_url = input("Enter the YouTube video URL: ")
generate_video_tags(youtube_url)
```

```python

import cv2
import numpy as np
import yt_dlp  # Using yt-dlp instead of pytube
import re
from transformers import pipeline

# Function to download video from a YouTube URL using yt-dlp
def download_video(video_url):
    try:
        ydl_opts = {
            'format': 'bestvideo+bestaudio/best',
            'outtmpl': 'downloaded_video.mp4',  # Output filename
            'noplaylist': True,  # Download single video
        }
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.extract_info(video_url, download=True)
        print("Downloaded video: downloaded_video.mp4")
    except Exception as e:
        print(f"Error downloading video: {e}")

# Function to extract video ID from YouTube URL
def extract_video_id(url):
    match = re.search(r"(?:(?<=v=)|(?<=youtu\.be/)|(?<=embed/)|(?<=watch\?v=)|(?<=/videos/)|(?<=/v/))([a-zA-Z0-9_-]{11})", url)
    if match:
        return match.group(1)
    else:
        raise ValueError("Invalid YouTube URL format")

# Function to apply YOLO object detection on video frames
def detect_events(video_path):
    # Load YOLO model and configuration
    net = cv2.dnn.readNet("/content/sample_data/yolov3.weights", "/content/sample_data/yolov3.cfg")
    layer_names = net.getLayerNames()
    output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]

    # Load video
    cap = cv2.VideoCapture(video_path)
    highlights = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Prepare frame for YOLO
        blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
        net.setInput(blob)
        outputs = net.forward(output_layers)

        # Process the outputs to extract detected objects
        for output in outputs:
            for detection in output:
                scores = detection[5:]
                class_id = np.argmax(scores)
                confidence = scores[class_id]

                if confidence > 0.5:  # Threshold for detection
                    highlights.append(frame)  # Save frame as highlight

    cap.release()
    return highlights

# Function to summarize video highlights using an attention-based model
def summarize_highlights(highlights):
    summarizer = pipeline("summarization")
    highlight_text = " ".join([f"Highlight frame {i+1}" for i in range(len(highlights))])  # Placeholder for actual highlight text

    # Summarize using attention model
    summary = summarizer(highlight_text, max_length=50, min_length=25, do_sample=False)
    return summary[0]['summary_text']

# Main function to run event detection and highlight extraction
def context_aware_event_detection(video_url):
    # Step 1: Download video
    download_video(video_url)

    # Step 2: Detect events
    highlights = detect_events("downloaded_video.mp4")

    # Step 3: Summarize highlights
    summary = summarize_highlights(highlights)

    print("Event Highlights Summary:")
    print(summary)

# Example usage
if __name__ == "__main__":
    video_url = input("Enter YouTube video URL: ")  # User inputs video URL
    context_aware_event_detection(video_url)
```

#Explore Transfer Learning

```python
import cv2
import numpy as np
import yt_dlp  # Use yt-dlp instead of pytube
from gtts import gTTS
import os
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.models import Model

# Function to download video from a YouTube URL using yt-dlp
def download_video(video_url):
    try:
        ydl_opts = {
            'format': 'bestvideo+bestaudio/best',
            'outtmpl': 'downloaded_video.mp4',
            'noplaylist': True,  # Only download single video
        }
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info_dict = ydl.extract_info(video_url, download=True)
            title = info_dict.get('title', None)  # Get video title
            print(f"Downloaded video: {title}")
            return title
    except Exception as e:
        print(f"Error downloading video: {e}")
        return None

# Function to save video title as audio
def save_title_as_audio(title):
    tts = gTTS(text=title, lang='en')
    audio_file_name = f"{title}.mp3".replace(" ", "_")  # Replace spaces with underscores
    tts.save(audio_file_name)  # Save the audio file
    print(f"Audio saved as {audio_file_name}")

# Function to extract frames from the video
def extract_frames(video_path, num_frames=30):
    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    interval = max(1, frame_count // num_frames)  # Ensure interval is at least 1

    frames = []
    current_frame = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if current_frame % interval == 0:
            frames.append(frame)
        current_frame += 1

    cap.release()
    return frames

# Function to extract features using a pre-trained model
def extract_features(frames):
    # Load MobileNetV2 model pre-trained on ImageNet
    base_model = MobileNetV2(weights='imagenet', include_top=False, pooling='avg')
    feature_model = Model(inputs=base_model.input, outputs=base_model.output)

    # Preprocess frames and extract features
    features = []
    for frame in frames:
        frame_resized = cv2.resize(frame, (224, 224))  # Resize to MobileNetV2 input size
        frame_preprocessed = preprocess_input(frame_resized)
        frame_expanded = np.expand_dims(frame_preprocessed, axis=0)
        feature = feature_model.predict(frame_expanded)
        features.append(feature)

    # Convert to numpy array and handle dimension issues
    features_array = np.array(features)
    if features_array.ndim > 1:  # Check if the array has more than 1 dimension
        return features_array.squeeze(axis=1)  # Squeeze if needed
    else:
        return features_array  # Return as is if it's already 1D

# Function to summarize the video by selecting key frames based on features
def summarize_video(features, num_summary_frames=5):
    # Check if features is empty
    if len(features) == 0:
        print("No features extracted; cannot summarize video.")
        return []

    # For simplicity, select frames with the maximum features
    norms = np.linalg.norm(features, axis=1, keepdims=True)  # Ensure norms are calculated correctly
    frame_indices = np.argsort(norms.flatten())[-num_summary_frames:]  # Flatten and sort indices
    return frame_indices

# Main function to run the transfer learning video summarization
def transfer_learning_video_summary(video_url):
    # Step 1: Download video
    video_title = download_video(video_url)
    if video_title is None:
        return  # Exit if video download failed

    # Step 2: Save video title as audio
    save_title_as_audio(video_title)

    # Step 3: Extract frames from video
    frames = extract_frames("downloaded_video.mp4")

    # Step 4: Extract features from frames using pre-trained model
    features = extract_features(frames)

    # Step 5: Summarize video by selecting key frames
    summary_indices = summarize_video(features)

# Example usage
if __name__ == "__main__":
    video_url = input("Enter YouTube video URL: ")  # User inputs video URL
    transfer_learning_video_summary(video_url)
```

#Establish Robust Evaluation Metrics
    Develop and utilize comprehensive evaluation metrics, including temporal consistency and user satisfaction, to assess the performance of the summarization algorithms and ensure they meet user expectations and industry standards.

```python
import cv2
import numpy as np
import yt_dlp  # Use yt-dlp instead of pytube
import os
import re

# Function to download video from a YouTube URL using yt-dlp
def download_video(video_url):
    try:
        ydl_opts = {
            'format': 'bestvideo+bestaudio/best',
            'outtmpl': 'downloaded_video.mp4',
            'noplaylist': True,  # Only download single video
        }
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.extract_info(video_url, download=True)
        print("Downloaded video: downloaded_video.mp4")
    except Exception as e:
        print(f"Error downloading video: {e}")

# Function to extract video ID from YouTube URL
def extract_video_id(url):
    match = re.search(r"(?:(?<=v=)|(?<=youtu\.be/)|(?<=embed/)|(?<=watch\?v=)|(?<=/videos/)|(?<=/v/))([a-zA-Z0-9_-]{11})", url)
    if match:
        return match.group(1)
    else:
        raise ValueError("Invalid YouTube URL format")

# Function to summarize the video by extracting key frames
def summarize_video(video_path, num_frames=5):
    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    interval = frame_count // num_frames if num_frames > 0 else 1  # Avoid division by zero

    key_frames = []
    current_frame = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if current_frame % interval == 0:
            key_frames.append(frame)
        current_frame += 1

    cap.release()
    return key_frames

# Function to evaluate temporal consistency (basic evaluation)
def evaluate_temporal_consistency(key_frames):
    if len(key_frames) < 2:
        return 0  # Not enough frames to evaluate

    consistency_scores = []
    for i in range(1, len(key_frames)):
        # Calculate difference between consecutive frames
        diff = cv2.absdiff(key_frames[i-1], key_frames[i])
        consistency_score = np.sum(diff) / (key_frames[i].size)  # Normalized score
        consistency_scores.append(consistency_score)

    # Average score for temporal consistency
    return np.mean(consistency_scores)

# Function to mock user satisfaction
def evaluate_user_satisfaction():
    # This could be a more complex evaluation in a real scenario
    # For demonstration, let's return a mock score
    return np.random.uniform(0, 10)  # Random user satisfaction score between 0 and 10

# Main function to run video summarization and evaluation
def context_aware_video_summary(video_url):
    # Step 1: Download video
    download_video(video_url)

    # Step 2: Summarize video
    key_frames = summarize_video("downloaded_video.mp4")

    # Step 3: Evaluate performance
    temporal_consistency = evaluate_temporal_consistency(key_frames)
    user_satisfaction = evaluate_user_satisfaction()

    print(f"Temporal Consistency Score: {temporal_consistency}")
    print(f"User Satisfaction Score: {user_satisfaction}")

# Example usage
if __name__ == "__main__":
    video_url = input("Enter YouTube video URL: ")  # User inputs video URL
    context_aware_video_summary(video_url)
```

#Version Control
    This program tracks versioned summaries with timestamps, lists AI learning resources, and fetches the latest summary for a user-selected topic.


```python
import datetime

# Dictionary to store versioned summaries
versioned_summaries = {}

def track_summary_versions(video_id, summary):
    """Track versions of summaries with timestamps."""
    if video_id not in versioned_summaries:
        versioned_summaries[video_id] = []
    # Append a tuple of (timestamp, summary) to keep track of when it was added
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    versioned_summaries[video_id].append((timestamp, summary))

def get_summaries(video_id):
    """Fetch all summaries for a given video ID."""
    return versioned_summaries.get(video_id, [])

def get_latest_summary(video_id):
    """Fetch the latest summary along with its timestamp for a given video ID."""
    summaries = versioned_summaries.get(video_id, [])
    return summaries[-1] if summaries else None

# Dictionary with learning resources
learning_resources = {
    "AI Bias and Fairness": ["https://youtu.be/Qul6F_Qoeq4"],
    "Explainable AI (XAI)": ["https://youtu.be/aovT0B8AJkA", "https://youtu.be/YJAXYpO9XYo"],
    "Generative Adversarial Networks (GANs)": ["https://youtu.be/TpMIssRdhco"],
    "Natural Language Processing (NLP)": ["https://youtu.be/5sLYAQS9sWQ"],
    "Computer Vision": ["https://youtu.be/ag3DLKsl2vk"],
}

def format_topic_to_id(topic):
    """Convert topic names to lowercase IDs with underscores, removing special characters."""
    return topic.replace(" ", "_").replace("(", "").replace(")", "").lower()

# Function to print all learning resources for AI topics
def print_learning_resources():
    print("Exploring Key Topics in Artificial Intelligence:")
    for topic in learning_resources.keys():
        print(f"- {topic}")

# Function to fetch resources and latest summaries for a specific topic
def get_learning_resources(topic):
    """Display learning resources and latest summary for the given topic."""
    video_id = format_topic_to_id(topic)
    resources = learning_resources.get(topic, [])

    if resources:
        print(f"\nResources for '{topic}':")
        for link in resources:
            print(f"- {link}")

    else:
        print(f"No resources found for the topic '{topic}'.")

# Example usage
print_learning_resources()

# Prompt user for a topic to display resources and summaries
user_topic = input("\nEnter the topic you want resources for: \n")
get_learning_resources(user_topic)

# Adding summaries for different topics with timestamps
track_summary_versions("ai_bias_and_fairness", "Summary of AI Bias and Fairness video - version 1")
track_summary_versions("ai_bias_and_fairness", "Summary of AI Bias and Fairness video - version 2")
track_summary_versions("explainable_ai_xai", "Summary of XAI video - version 1")
track_summary_versions("generative_adversarial_networks_gans", "Summary of GANs video - version 1")
track_summary_versions("natural_language_processing_nlp", "Summary of NLP video - version 1")
track_summary_versions("computer_vision", "Summary of Computer Vision video - version 1")

# Fetch the latest summary for the topic entered by the user
latest_summary = get_latest_summary(format_topic_to_id(user_topic))
if latest_summary:
    timestamp, summary = latest_summary
    print(f"\nLatest summary for '{user_topic}': ({timestamp}, '{summary}')")
else:
    print(f"No summaries found for '{user_topic}'.")
```
